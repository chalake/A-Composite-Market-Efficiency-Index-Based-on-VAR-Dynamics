```python
import pandas as pd
from pathlib import Path

# 定义文件夹路径
folder_path = r'C:\Users\yzm\Downloads\data_玉米(现货价&期货价)'

# 定义目标月份
target_months = {
    '2405': [3, 4],
    '2407': [5, 6],
    '2409': [7, 8],
    '2411': [9, 10],
    '2501': [11, 12],
    '2503': [1, 2]
}

# 创建一个空的DataFrame来存储合成数据
synthetic_data = pd.DataFrame()

# 遍历文件夹中的每个Excel文件
for file_path in Path(folder_path).glob('*.xlsx'):
    # 读取Excel文件
    df = pd.read_excel(file_path)
    
    # 获取文件名中的合约号
    contract_number = file_path.stem.split('合约')[0][-4:]
    
    # 获取目标月份
    target_month = target_months.get(contract_number)
    
    if target_month:
        # 筛选出目标月份的数据
        filtered_data = df[df['日期'].dt.month.isin(target_month)]
        
        # 添加到合成数据中
        synthetic_data = pd.concat([synthetic_data, filtered_data])

# 将合成数据保存到新的Excel文件中
synthetic_data.to_excel(r'C:\Users\yzm\Downloads\synthetic_data.xlsx', index=False)
```
