```python
import pandas as pd

# 读取原始数据
file_path = 'D:/统计建模/综合绩效.xlsx'
df = pd.read_excel(file_path)

# 将日期列设置为索引
df.set_index('日期', inplace=True)

# 计算每年的绩效年均值
yearly_mean = df.resample('Y').mean()

# 保存结果到新文件
output_file = 'D:/统计建模/年均综合绩效.xlsx'
yearly_mean.to_excel(output_file)
```

```python
import pandas as pd

# 读取原始数据
file_path = 'D:/统计建模/综合绩效.xlsx'
df = pd.read_excel(file_path)

# 将日期列设置为索引
df.set_index('日期', inplace=True)

# 计算每个季度的绩效季度均值
quarterly_mean = df.resample('Q').mean()

# 保存结果到新文件
output_file = 'D:/统计建模/季度均综合绩效.xlsx'
quarterly_mean.to_excel(output_file)
```
