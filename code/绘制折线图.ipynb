```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False  # 
plt.rcParams['font.size'] = 16

# Excel
file_path = 'D://.xlsx'
df = pd.read_excel(file_path,sheet_name='')

# 
dates = df.iloc[:, 0]  # 
prices = df.iloc[:, 1:]  # 

# 
colors =['#F9F3D1', '#D4E537', '#B3BE1A', '#989400', '#2C2B19', '#A4934D', '#BCA67E', '#D1BD8C', '#E8CC99', '#7F7C84', '#D67573', '#F4C7A8', '#F6E7CE', '#FCFAF1', '#BFB636', '#F2C572', '#F27405', '#F25C05', '#F23207', '#D7DBE4', '#B9C5D1', '#9BB2D2']
plt.figure(figsize=(16, 9),dpi=300)
for i, col in enumerate(prices.columns):
    plt.plot(dates, prices[col], label=col, color=colors[i],linewidth=4)

# 
plt.xlabel('')
plt.ylabel('')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)  # 

# 
plt.show() 
```
