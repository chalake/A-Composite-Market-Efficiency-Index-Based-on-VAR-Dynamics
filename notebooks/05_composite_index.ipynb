```python
import pandas as pd
from src.indexing.weights import compute_entropy_weights
from src.indexing.composite import build_entropy_weighted_index

# Load features
features = pd.read_excel(
    "data/zscore_standardized_features_shifted.xlsx"
)

# Compute weights
weights = compute_entropy_weights(
    df=features,
    index_col="window"
)

# Build composite index
market_efficiency_index = build_entropy_weighted_index(
    features=features,
    weights=weights,
    index_col="window",
    index_name="market_efficiency_index"
)

market_efficiency_index.to_excel(
    "data/composite_market_efficiency_index.xlsx"
)
```
